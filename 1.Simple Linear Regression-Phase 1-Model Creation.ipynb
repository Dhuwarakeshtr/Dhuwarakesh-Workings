```python
#importing the "pandas" to read the tabel sets, data sets, files are in the "csv" format | And name the "pandas" as "pd" to use in short-form in further process.
import pandas as pd
```

```python
#assigning the csv file datas to the variable "datasets"
#syntax --> variable_name=pd.read_csv("filename.csv")

# Model Creation - Data Preprocessing

datasets=pd.read_csv("Salary_Structures.csv")
```

```python
# Below we can see the stored memory of that variable - "datasets"
datasets
```

```python
# assigning the 1st column "Years of Experience" as input/inpendent variable
# Model Creation - input data split

independent=datasets[["Years of Experience"]]
independent
```

```python
# assigning the 2nd column "Salary" as output/dependent variable
# Model Creation - output data split

dependent=datasets[["Salary"]]
dependent
```

```python
# Model Creation - Split train and test data sets
# Using from "sklearn.model_selection" import "train_test_split" for the "Machine Learning Concept"
#x_train(i/p train data position),x_test(i/p test data position),y_train(o/p train data position),y_test(o/p test data position)=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# For the "SLR" algorithm in "ML" we need to use file name - "sklearn".linear_model and use class/function name - "LinearRegression".
# And assign that "class or function name to a variable "regressor".
# The above 2 points are the "Procedures" in the "Model Creation" stage.
# Learning Stage is below.
# "regressor.fit(X-train,y_train) - stands for the formula in that SLR-algorithm. i.e., Y=Wx+b. (this is used to find out the weight and bias value from the training datasts).
# And once we ge the weight (w) & bias (b) values which we will apply in the formual. Model/AI Brain is created.

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
weight=regressor.coef_
weight
```

```python
bias=regressor.intercept_
bias
```

```python
# Belwo is the process to test test the created Model/AI brain using the split test datasets.
# So, we are try to get the "predicted" value using the split test data we have splited before.

y_pred=regressor.predict(x_test)
```

```python
# Belwo is the process for "Evaluation Metrics". using the R2(r_square) algorithm 

from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#The result vause is near to 1, the created Model/AI Brain is GOOD.
r_score
```
